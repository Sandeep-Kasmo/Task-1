```python
# if there are duplicates, you can drop them using
df=df.drop_duplicates(subset=['customer_id'],keep='first')
```

```python
df.info()
```

```python
'''we found that there are null values in email,
so we replace null values with name@email.com'''
```

```python
df['email']=df.apply(lambda row:f"{row['name'].lower().replace(' ','_')}@email.com" if pd.isnull(row['email']) else row['email'],axis=1)
```

```python
df.info()
```

```python
'''so we have replaced null emails, then we again see that phone number is also null'''
```

```python
df['phone']=df.apply(lambda row:'000-000-0000' if pd.isnull(row['phone']) else row['phone'],axis=1)
```

```python
df.info()
```

```python
#so we have replaced all null values, now lets check for incorrect data, like any incorrect emails
```

```python
df['email_check']=df['email'].astype(str).str.match(r'^[a-zA-Z0-9._]+@[a-z]+\.[a-z]{2,}$',na=False)
```

```python
df['email_check']
```

```python
df['email_check'].value_counts()
```

```python
#since all the flags are true, we can move further, but drop that extra flag column
df.drop(columns=['email_check'],inplace=True)
```

```python
df.info()
```

```python
# now lets do the same for phone
df['phone_check']=df['phone'].astype(str).str.match(r'^\d{3}-\d{3}-\d{4}$',na=False)
```

```python
df['phone_check'].value_counts()
```

```python
df.loc[df['phone_check'] == False, 'phone'] = df.loc[df['phone_check'] == False, 'phone'].astype(str).str.replace(r'[\(\)]', '', regex=True)
```
